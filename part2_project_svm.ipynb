Importing all the required packages

```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import GridSearchCV
import os
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA
```

```python
flat_data_arr=[]#er 
target_arr=[]
#please use datadir='/content' if the files are upload on to google collab
#else mount the drive and give path of the parent-folder containing all category images folders.
datadir='E:\\university\\hosh\\project-DecisionTree\\images'
Categories = ['Cars' ,  'Cricket ball']
for i in Categories:
  print(f'loading... category : {i}')
  path=os.path.join(datadir,i)
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img) , as_gray=True)
    img_resized=resize(img_array,(150,150,3))
    flat_data_arr.append(img_resized.flatten())
    target_arr.append(Categories.index(i))
  print(f'loaded category:{i} successfully')
flat_data=np.array(flat_data_arr)
target=np.array(target_arr)
df=pd.DataFrame(flat_data)
df['Target']=target
df
```

Splitting the data into training and testing data

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=77)
print('Splitted Successfully')
```

***PCA***

```python
pca = PCA(n_components = 0.95) # 95 % of the information is all of them! 
pca.fit(x_train)
X_train_pca = pca.transform(x_train)
X_test_pca = pca.transform(x_test)
X_train_pca_df = pd.DataFrame(X_train_pca)
X_test_pca_df = pd.DataFrame(X_test_pca)
X_train_pca_df
```

```python
X_test_pca_df
```

***first gird search on params***

```python
param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf','poly','linear']}
svc=svm.SVC(probability=True)
print("The training of the model is started, please wait for while as it may take few minutes to complete")
model=GridSearchCV(svc,param_grid)
model.fit(X_train_pca_df,y_train)
print('The Model is trained well with the given images')
model.best_params_
```

```python
param_grid={'C':[10],'gamma':[0.0001],'kernel':['rbf']}
svc=svm.SVC(probability=True)
model=GridSearchCV(svc,param_grid)
model.fit(X_train_pca_df,y_train)
```

***Double grid Search***

```python
param_grid_second = {'C':[ 10 , 15 , 20 , 5] , 'gamma':[0.00005 , 0.0001 ,0.00015 ,0.0002 ] , 'kernel':['rbf']}
svc_second=svm.SVC(probability=True)
model_grid_second = GridSearchCV(svc_second,param_grid_second)
model_grid_second.fit(X_train_pca_df, y_train)
model_grid_second.best_params_
```

```python
scores = cross_val_score(model,X_train_pca_df,y_train,cv=4)
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```

Predicting our testing data

```python
y_pred=model_grid_second.predict(X_test_pca_df)
print("The predicted Data is :")
y_pred
```
